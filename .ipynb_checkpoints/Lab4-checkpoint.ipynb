# TOP 5 TALKERS

```python
def reverse_lookup(ip):
    try:
        domain = socket.gethostbyaddr(ip)
    except Exception:
        domain = 'Reverse lookup unavailable'
    return domain

pd.src_ip = df.src_ip
values = src_ip.value_counts().keys().tolist() 
counts = src_ip.value_counts().tolist()
for x in range(5):
    print(x+1,')\tIP:',values[x],'\n\tCount: ',counts[x])
    print('\t',reverse_lookup(values[x]))
```

# TOP 5 LISTENERS

```python
dst_ip = df.dst_ip
values = dst_ip.value_counts().keys().tolist() 
counts = dst_ip.value_counts().tolist()
for x in range(5):
    print(x+1,')\tIP:',values[x],'\n\tCount: ',counts[x])
    print('\t',reverse_lookup(values[x]))
```
